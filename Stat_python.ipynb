```python
import numpy as np
def detect_outliers(data,threshold=2):
   mean=np.mean(data)
   std=np.std(data)
   outliers=[x for x in data if abs((x-mean)/std)>threshold]
   return outliers

#Example usage:
data=[10,12,11,13,10,11,12,11]
print("Outliers:", detect_outliers(data))

data1=[10,11,12,1009,2,19,13]
print("Outliers:",detect_outliers(data1))
```

```python
data=[1,2,3,4,5,6,7,8,9,90,1,2,3,4,5,6,7,8,9,1,2,3,4,5,6,7,8,9,1,2,3,4,5,6,7,8,9,1,2,3,4,5,6,7,8,9]
mean=np.mean(data)
std=np.std(data)
print("Mean of array1 is",mean)
print("STD of array1 is",std)

threshold=2
outliers=[]
for x_value in data:
    z_score=(x_value-mean)/std
    print("z_score is:",z_score)
    if np.abs(z_score)>threshold:
        outliers.append(x_value)
        
print("Outliers are :",outliers)       
        
        
```

```python
def detect_outliers_iqr(data):
    q1=np.percentile(data,25)
    q3=np.percentile(data,75)
    iqr=q3-q1
    lower_bound=q1-1.5*iqr
    upper_bound=q3+1.5*iqr
    return[x for x in data if x < lower_bound or x > upper_bound]

#Example usage:
print("Outliers (IQR):",detect_outliers_iqr(data))
```
